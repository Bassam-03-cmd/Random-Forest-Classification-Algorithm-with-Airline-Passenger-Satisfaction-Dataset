## Next step will be the feature engineering to decide the features that will be included in my study

### Firstly, I made the feature engineering using Recursive Feature Elimination (RFE) with the most 12 features that the RFE sees them as the most important features that affects in the classification accuracy. In addition, I will validate the model that I studied in the RFE 

```python
#Convert the dataset in both train and test to numeric data that the classifier can treat with easily
APS_train_data['Gender'] = APS_train_data['Gender'].replace({'Male': 0, 'Female': 1})
APS_train_data['Customer Type'] = APS_train_data['Customer Type'].replace({'Loyal Customer': 0, 'disloyal Customer': 1})
APS_train_data['Type of Travel'] = APS_train_data['Type of Travel'].replace({'Personal Travel': 0, 'Business travel': 1})
APS_train_data['Class'] = APS_train_data['Class'].replace({'Eco': 0, 'Eco Plus': 1, 'Business' : 2})
APS_train_data['satisfaction'] = APS_train_data['satisfaction'].replace({'neutral or dissatisfied': 1, 'satisfied' : 2})

APS_test_data['Gender'] = APS_test_data['Gender'].replace({'Male': 0, 'Female': 1})
APS_test_data['Customer Type'] = APS_test_data['Customer Type'].replace({'Loyal Customer': 0, 'disloyal Customer': 1})
APS_test_data['Type of Travel'] = APS_test_data['Type of Travel'].replace({'Personal Travel': 0, 'Business travel': 1})
APS_test_data['Class'] = APS_test_data['Class'].replace({'Eco': 0, 'Eco Plus': 1, 'Business' : 2})
APS_test_data['satisfaction'] = APS_test_data['satisfaction'].replace({'neutral or dissatisfied': 1, 'satisfied' : 2})
```

```python
#CHECK THE TEST DATASET
APS_test_data
```

```python
#CHECK THE TRAIN DATASET
APS_train_data
```

```python
test_of_FE = APS_train_data
X = test_of_FE.iloc[:, :-1]
Y = test_of_FE['satisfaction']
```

```python
model = RandomForestClassifier ()
rfe = RFE(estimator = model , n_features_to_select = 12)
X_rfe = rfe.fit_transform (X, Y)
```

```python
# Get the selected features
selected_features = [X.columns[i] for i in range(len(X.columns)) if rfe.support_ [i]]
print("Selected features: ", selected_features)
```

```python
#Drop the non-selected features to train the model with the selected features
columns_to_drop = ['Gender', 'Customer Type', 'Departure/Arrival time convenient', 'Gate location', 'Food and drink', 'Departure Delay in Minutes',
                  'Arrival Delay in Minutes', 'Checkin service', 'Inflight service', 'Baggage handling', 'Cleanliness']
test_of_FE = test_of_FE.drop(columns = columns_to_drop, axis = 1, inplace=False)
```

```python
#Set the new dataset that will be used for training the model with the selected features from RFE
X_train = test_of_FE.iloc[:, :-1]
Y_train = test_of_FE['satisfaction']
```

```python
#Train the random forest classifier
rf_clf = RandomForestClassifier()
rf_clf.fit(X_train, Y_train)
```

```python
#Get the testing dataset and drop the non-selected features as we did in the training dataset
dataset_with_rfe_features_for_test = APS_test_data
columns_to_drop = ['Gender', 'Customer Type', 'Departure/Arrival time convenient', 'Gate location', 'Food and drink', 'Departure Delay in Minutes',
                  'Arrival Delay in Minutes', 'Checkin service', 'Inflight service', 'Baggage handling', 'Cleanliness']
dataset_with_rfe_features_for_test = dataset_with_rfe_features_for_test.drop(columns = columns_to_drop, axis = 1, inplace=False)
```

```python
X_test = dataset_with_rfe_features_for_test.iloc[:, :-1]
Y_test = dataset_with_rfe_features_for_test['satisfaction']
```

```python
#Predict the labeles for the testing dataset 
Y_pred = rf_clf.predict(X_test)
```

```python
#For the evaluation of the RFE selected features model, I will use the metrices measurment and cross validation
clf_cross = RandomForestClassifier()
scores = cross_val_score(clf_cross, X_train, Y_train, cv = 10, scoring='accuracy')
scores
```

```python
clf_y_pred = cross_val_predict(clf_cross, X_test, Y_test, cv=10)
report = classification_report(Y_test, clf_y_pred)
print(report)
```

```python
cm = confusion_matrix(Y_test, Y_pred)
accuracy = accuracy_score(Y_test, Y_pred)
precision = precision_score(Y_test, Y_pred, average='weighted')
recall = recall_score(Y_test, Y_pred, average='weighted')
f1 = f1_score(Y_test, Y_pred, average='weighted')
print("Confusion Matrix for the measurment of classification:")
print(cm)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
